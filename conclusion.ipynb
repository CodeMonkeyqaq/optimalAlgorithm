```python
import numpy as np
from matplotlib import pyplot as plt
from test_function import *
```

```python
a = np.load('fun.npz')
fun = a['fun']
```

```python
np.shape(fun)
```

```python
max_arr = np.zeros((10,30,6))
min_arr = np.zeros((10,30,6))
mean_arr = np.zeros((10,30,6))
for i in range(10):
    for j in range(30):
        for p in range(6):
            max_arr[i][j][p] = max(fun[i][j][p])
            min_arr[i][j][p] = min(fun[i][j][p])
            mean_arr[i][j][p] = np.mean(fun[i][j][p])
        
```

```python
sum = 0
for i in range(100):
    sum += fun[0][0][1][i]
sum/30
```

```python
de_arr = np.zeros((10,30))
abc_arr = np.zeros((10,30))
cs_arr = np.zeros((10,30))
pso_arr = np.zeros((10,30))
sa_arr = np.zeros((10,30))
csa_arr = np.zeros((10,30))
for i in range(10):
    for j in range(30):
        de_arr[i][j] = (mean_arr[i][j][0])
        abc_arr[i][j] = (mean_arr[i][j][1])
        cs_arr[i][j] = (mean_arr[i][j][2])
        pso_arr[i][j] = (mean_arr[i][j][3])
        sa_arr[i][j] = (mean_arr[i][j][4])
        csa_arr[i][j] = (mean_arr[i][j][5])
```

```python
plt.figure()
plt.plot(de_arr[0], label="de", linestyle="-")
plt.plot(abc_arr[0], label="abc", linestyle="--")
plt.plot(cs_arr[0], label="cs", linestyle=":")
plt.plot(pso_arr[0], label="pso", linestyle="-")
plt.plot(sa_arr[0], label="sa", linestyle="-.")
plt.plot(csa_arr[0], label="csa", linestyle=":")
plt.legend()
plt.title(test_function_names[0])
plt.xlabel("dim")
plt.ylabel("function_value")
plt.show()
```
