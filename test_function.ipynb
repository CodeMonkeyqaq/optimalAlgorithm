```python
from matplotlib import pyplot as plt  
import numpy as np  
import math
import os
from mpl_toolkits.mplot3d import Axes3D  
```

```python
def draw_pic(X,Y,Z,title):
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.plot_surface(X, Y, Z, cmap=plt.cm.hot)
    ax.set_title(title)
    plt.show()
```

```python
# 多局部最优值多维测试函数
```

```python
# fx=0 在 x=(0,0,...,0)处取得
```

```python
def Ackley():
    a = 20
    b = 0.2 
    c = 2 * np.pi
    x = np.arange(-32.768, 32.768,0.1) #创建等差数组 与np.linspace类似但不相同（arange使用的是步长不是步数） 
    y = np.arange(-32.768, 32.768, 0.1)  
    X, Y = np.meshgrid(x, y)
    d = 2
    Z = -a * np.exp(-b*np.sqrt(1/d*(X**2+Y**2)))-np.exp(1/d*(np.cos(c*X)+np.cos(c*Y)))+a+np.exp(1)
    return X, Y, Z, "Ackley function"
```

```python
# fx=0 在 x=(0,0,...,0)处取得
```

```python
def Griewank():
    x = np.arange(-600, 600, 0.1) #创建等差数组 与np.linspace类似但不相同（arange使用的是步长不是步数） 
    y = np.arange(-600, 600, 0.1)  
    X, Y = np.meshgrid(x, y)
    d = 2
    Z = X**2/4000+Y**2/4000-np.cos(X/np.sqrt(1))-np.cos(Y/np.sqrt(2))+1
    return X, Y, Z, "Griewank function"
```

```python
# fx=0 在 x=(1,1,...,1)处取得
```

```python
def LevyF():
    x = np.arange(-10, 10, 0.1) #创建等差数组 与np.linspace类似但不相同（arange使用的是步长不是步数） 
    y = np.arange(-10, 10, 0.1)  
    X, Y = np.meshgrid(x, y)
    w1 = 1+(X-1)/4
    w2 = 1+(Y-1)/4
    d = 2
    Z = np.sin(np.pi*w1)**2+(w1-1)**2*(1+10*np.sin(np.pi*w1+1)**2)+(w2-1)**2*(1+np.sin(2*np.pi*w2)**2)
    return X, Y, Z, "Levy function"
```

```python
# fx=0 在 x=(420.9687,420.9687,...,420.9687)处取得
```

```python
def Schwefel():
    x = np.arange(-500, 500, 1) #创建等差数组 与np.linspace类似但不相同（arange使用的是步长不是步数） 
    y = np.arange(-500, 500, 1)  
    X, Y = np.meshgrid(x, y)
    d = 2
    Z = 418.9828*d-(X*np.sin(np.sqrt(np.abs(X))))-(Y*np.sin(np.sqrt(np.abs(Y))))
    return X, Y, Z, "Schwefel function"
```

```python
# fx=0 在 x=(0,0,...,0)处取得
```

```python
def Rastrigin():
    x = np.arange(-5.12, 5.12, 0.1) #创建等差数组 与np.linspace类似但不相同（arange使用的是步长不是步数） 
    y = np.arange(-5.12, 5.12, 0.1)  
    X, Y = np.meshgrid(x, y)
    d = 2
    Z = 10 * d + X ** 2 - 10 * np.cos(2 * np.pi * X) + Y ** 2 - 10 * np.cos(2 * np.pi * Y)
    return X, Y, Z, "Rastrigin function"
```

```python
# 碗状函数，连续凸单峰
```

```python
# fx=0 在 x=(0,0,...,0)处取得
```

```python
def Sphere():
    x = np.arange(-5.12, 5.12, 0.1) #创建等差数组 与np.linspace类似但不相同（arange使用的是步长不是步数） 
    y = np.arange(-5.12, 5.12, 0.1)  
    X, Y = np.meshgrid(x, y)
    d = 2
    Z = X**2+Y**2
    return X, Y, Z, "Sphere function"
```

```python
# 盘状函数 最优值唯一，且下降缓慢
```

```python
# fx=0 在 x=(0,0,...,0)处取得
```

```python
def Zakharov():
    x = np.arange(-5.12, 5.12, 0.1) #创建等差数组 与np.linspace类似但不相同（arange使用的是步长不是步数） 
    y = np.arange(-5.12, 5.12, 0.1)  
    X, Y = np.meshgrid(x, y)
    d = 2
    Z = X**2+Y**2+(0.5*X+Y)**2+(0.5*X+Y)**4
    return X, Y, Z, "Zakharov function"
```

```python
# 山谷型函数，全局最优值位于山谷中，难以收敛
```

```python
# fx=0 在 x=(1,1,...,1)处取得
```

```python
def Rosenbrock():
    x = np.arange(-5, 10, 0.1) #创建等差数组 与np.linspace类似但不相同（arange使用的是步长不是步数） 
    y = np.arange(-5, 10, 0.1)  
    X, Y = np.meshgrid(x, y)
    d = 2
    Z = 100*(Y-X**2)**2+(X-1)**2
    return X, Y, Z, "Rosenbrock function"
```

```python
# 陡峭山脊型函数，函数多峰，局部最小值下降迅速
```

```python
# d=2时 fx=-1.8013 在x=(2.2 1.57)处，其余不同
```

```python
def Michalewicz():
    x = np.arange(0, np.pi, 0.1) #创建等差数组 与np.linspace类似但不相同（arange使用的是步长不是步数） 
    y = np.arange(0, np.pi, 0.1)  
    X, Y = np.meshgrid(x, y)
    d = 2
    Z = -(np.sin(X)*np.sin(X**2/np.pi)**20+np.sin(Y)*np.sin(2*Y**2/np.pi)**20)
    return X, Y, Z, "Michalewicz function"
```

```python
# fx=-39.16599*d 在x=(-2.903534,-2.903534,...,-2.903534)处取得
```

```python
def Styblinski_Tang():
    x = np.arange(-5, 5, 0.1) #创建等差数组 与np.linspace类似但不相同（arange使用的是步长不是步数） 
    y = np.arange(-5, 5, 0.1)  
    X, Y = np.meshgrid(x, y)
    d = 2
    Z = 1/2*(X**4-16*X**2+5*X+Y**4-16*Y**2+5*Y)
    return X, Y, Z, "Styblinski_Tang function"
```
